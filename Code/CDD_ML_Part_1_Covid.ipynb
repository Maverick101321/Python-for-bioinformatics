## **Importing libraries**

```python
# Import necessary libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

## **Search for Target protein**

### **Target search for Coronavirus**

```python
# Target search for coronavirus
target = new_client.target #defining the target var(x)
target_query = target.search('coronavirus') #defining the target query var(x)
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
# Target search for brain cancer exploring...
target2 = new_client.target #defining the target var(x)
target_query2 = target.search('brain cancer') #defining the target query var(x)
targets2 = pd.DataFrame.from_dict(target_query2)
targets2
```

### **Select and retrieve bioactivity data for *SARS coronavirus 3C like proteinase 6th entry 7'th number from start**

We will assign the sixth entry (which corresponds to the target protein, *SARS coronavirus 3C like proteinase*) to the ***selected_target*** variable 

```python
selected_target = targets.target_chembl_id[6]
selected_target
```

Here, we will retrieve only bioactivity data for the selected *SARS coronavirus 3C like proteinase*

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
print(res)
```

```python
df = pd.DataFrame.from_dict(res)
```

## **Handling missing data**
If any compounds has missing value for the **standard_value** and **canonical_smiles** column then dropping it.

```python
df2 = df[df.standard_value.notna()] #filtering standard_value and keeping only the not null
df2 = df2[df.canonical_smiles.notna()] #filtering canonical_smiles and keeping only the not null
df2
```

```python
len(df2.canonical_smiles.unique())
```

```python
#Getting rid of any duplicates in the canonical_smiles column
df2_nr = df2.drop_duplicates(['canonical_smiles'])
df2_nr
```

```python
len(df2_nr.canonical_smiles.unique())
```

## **Data pre-processing of the bioactivity data**

### **Combine the 3 columns (molecule_chembl_id,canonical_smiles,standard_value) and bioactivity_class into a DataFrame**

```python
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2_nr[selection]
df3
```

Saves dataframe to CSV file

```python
df3.to_csv('SARS_coronavirus_3C_proteinase_bioactivity_data_preprocessed.csv', index=False)
```

### **Labeling compounds as either being active, inactive or intermediate**
The bioactivity data is in the IC50 unit. Compounds having values of less than 1000 nM will be considered to be **active** while those greater than 10,000 nM will be considered to be **inactive**. As for those values in between 1,000 and 10,000 nM will be referred to as **intermediate**. 

```python
df4 = pd.read_csv('SARS_coronavirus_3C_proteinase_bioactivity_data_preprocessed.csv')
```

```python
df4
```

```python
bioactivity_threshold = []
for i in df4.standard_value:
  if float(i) >= 10000:
    bioactivity_threshold.append("inactive")
  elif float(i) <= 1000:
    bioactivity_threshold.append("active")
  else:
    bioactivity_threshold.append("intermediate")
```

```python
bioactivity_class = pd.Series(bioactivity_threshold, name='class')
df5 = pd.concat([df4, bioactivity_class], axis=1)
df5
```

Saves dataframe to CSV file

```python
df5.to_csv('SARS_coronavirus_3C_proteinase_bioactivity_data_curated.csv', index=False)
```
